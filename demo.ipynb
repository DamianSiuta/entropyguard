# ðŸ›¡ï¸ EntropyGuard Demo

This notebook demonstrates the capabilities of the EntropyGuard engine: removing PII, stripping HTML, and semantic deduplication.

```python
import polars as pl
import matplotlib.pyplot as plt
from entropyguard.cli.pipeline import Pipeline
```

## 1. Load & Inspect Raw Data

We load the generated stress test data (containing HTML tags and duplicates).

```python
df_raw = pl.read_ndjson('stress_test.jsonl')
print(f'Raw rows: {len(df_raw)}')
print(df_raw.head(5))
```

## 2. Run EntropyGuard Pipeline

We run the full pipeline with Sanitization and Semantic Deduplication (Threshold: 0.85).

```python
pipeline = Pipeline()
input_path = 'stress_test.jsonl'
output_path = 'demo_notebook_output.jsonl'
text_column = 'text'
min_length = 10
dedup_threshold = 0.85

result = pipeline.run(
    input_path=input_path,
    output_path=output_path,
    text_column=text_column,
    required_columns=None,
    min_length=min_length,
    dedup_threshold=dedup_threshold,
)
print('Pipeline result:', result)

df_clean = pl.read_ndjson(output_path)
print(f'Clean rows: {len(df_clean)}')
df_clean.head(5)
```

## 3. Results Analysis

Visualize the reduction in noise and redundancy.

```python
before = 10000
after = len(df_clean)

plt.figure(figsize=(6, 4))
plt.bar(['Before', 'After'], [before, after], color=['gray', 'green'])
plt.ylabel('Number of rows')
plt.title('EntropyGuard Noise & Redundancy Reduction')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
